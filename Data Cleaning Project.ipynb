## Data Cleaning : Process of detecting, correcting, handling errors in raw data to make it suitable for analysis or modelling 
#### Data is collected from multiple sources such as web scraping, databases, APIs, and surveys.
#### All collected information is stored as raw data for further cleaning and analysis.
#### Raw data often contains issues like missing values, duplicate records, spelling errors, outliers, incorrect data types, and inconsistent formatting.
#### These problems must be identified and resolved during data cleaning to ensure accurate analysis.
## Step By Step Data Cleaning Process :
#### Data collection and inspection involve understanding the type of data in each column, such as numerical, categorical, date-time, or text.
#### This step also focuses on identifying missing or duplicate values, incorrect data types, and inconsistencies in the dataset.
## Data cleaning techniques 
#### It include using summary statistics, distribution plots, and dataset info to understand data types and non-null counts.
#### Frequency analysis for categorical data and visual inspection help identify patterns, anomalies, and issues in the dataset.

## Handling Missing Values
### Delete rows / columns
#### → When more than 50% of data is missing
### Imputation → Filling missing values
#### Mean / Median / Mode → Numerical columns
#### Mode / “Unknown” → Categorical columns
#### Forward fill / Backward fill → Date-time columns & numerical columns
### Predictive Imputation using ML model
### Example:
#### Price column → 15% data missing
#### 85% values present
#### Use available data as features to predict missing values

inplace=False (Default): The method returns a new DataFrame with the modifications, leaving the original DataFrame unchanged. This is generally the recommended approach as it promotes safer, more readable code and allows for method chaining.

inplace=True: The operation modifies the original DataFrame directly. The method returns None. 

```python
import re
df['Age'] = df['Age'].astype(str).str.extract(r'(\d+)').astype(float)
```

```python
df['Age'].unique()
```

```python
age_median = df['Age'].median()
age_median
```

```python
df['Age'] = df['Age'].fillna(age_median).astype('int64')
```

```python
df['Customer_ID'] = 'CUST' + df['Customer_ID'].str[4:]
```

```python
df[df['Customer_ID'].duplicated()]
```

```python
# df = df[~df['Customer_ID'].duplicated()]
df = df.drop_duplicates(subset='Customer_ID', keep = 'first')
```

```python
df['Gender'] = df['Gender'].replace({'male' : 'Male','M' : 'Male','female' : 'Female', 'F' : 'Female'})
```

```python
df['Gender'] = df['Gender'].fillna('Prefer Not To Say')
```

```python
df['City'] = df['City'].str.upper()
```

```python
Feedback_Score_mode = df['Feedback_Score'].mode()[0] # for categorical values
# Feedback_Score_median = df['Feedback_Score'].median() -- for numerical values
Feedback_Score_mode
```

```python
df['Feedback_Score'] = pd.to_numeric(df['Feedback_Score'], errors='coerce')
```

```python
df['Feedback_Score'] = df['Feedback_Score'].fillna(Feedback_Score_mode)
```

```python
df['Feedback_Score'].unique()
```

```python
df['Feedback_Score'].value_counts()
```

```python
df['Feedback_Score'].isnull().sum()
```

```python
Purchase_Amount_avg = df['Purchase_Amount'].median()
Purchase_Amount_avg
```

```python
df['Purchase_Amount'] = df['Purchase_Amount'].fillna(Purchase_Amount_avg)
```

```python
df.loc[pd.to_numeric(df['Phone_Number'], errors='coerce') < 0, 'Phone_Number'] = 'Unknown'
```

```python
df['Phone_Number'] = df['Phone_Number'].astype(str).str.replace(r'\D', '', regex=True).str[-10:]
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('' , 'Unknown')
```

```python
Q1 = df['Purchase_Amount'].quantile(0.25)
Q3 = df['Purchase_Amount'].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
Upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound
```

```python
Upper_bound
```

```python
df_clean = df[
    (df['Purchase_Amount'] >= lower_bound) &
    (df['Purchase_Amount'] <= Upper_bound)
]
```
